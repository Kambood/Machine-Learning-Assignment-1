```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([2,0,-1,1,-2,2])
y = np.array([2,0,1,-1,0,0])
wStar = 0.1
b = 0.3
yhat = wStar*x + b
```

```python
plt.figure(figsize=(10,7))
plt.title("Q3.b",fontsize=25)
plt.xlabel("x",fontsize=18)
plt.ylabel("y", fontsize=18)
plt.scatter(x,y, color="red")
plt.plot(x,yhat, linestyle="dotted", marker='o')
```
